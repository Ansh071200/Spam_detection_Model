```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spam.csv")
```

```python
df.head()
```

```python
df.Category.value_counts()
```

```python
# new column to get numeric value for the spam and non spam emails based on the value of category
df["Numeric_Spam"]=df.Category.apply(lambda x : 1 if x == "spam" else 0)
```

```python
 from sklearn.model_selection import train_test_split
```

```python
# message body is x and whether its spam or not its y 
# with x as dependent variable and y as independent variable
# splitting dataset into two for unbiased results

X_train, X_test,y_train,y_test = train_test_split(df.Message,df.Numeric_Spam, test_size= 0.2)


# 20 percent in training and rest in the testing
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
# gives an numpy array
X_train.values
```

```python
# transforms the dataset into BOW vector with a function as the numpy array
X_train_cv = v.fit_transform(X_train.values)
```

```python
# a generated sparse matrix on the basis of BOW approach
X_train_cv.toarray()[:2]
```

```python

# for training the model
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_cv,y_train)
```

```python
X_test_cv = v.transform(X_test)
X_test_cv.toarray()
```

```python

from sklearn.metrics import classification_report
# predict the X_test_cv model
y_pred = model.predict(X_test_cv)

# generate a report on the basis of the prediction, accuracy of the model and the dataset
print(classification_report(y_test,y_pred))
```

```python

# give sample emails
emails = [
    "Free Entry in 2 a wkly "
]

# convert it into a vector using vectorizer
email_check = v.transform(emails) 
# check through the model whether the email is spam or not
model.predict(email_check)

```
